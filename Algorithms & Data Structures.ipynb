# 3. Three types of asymptotic notations __[Notation](https://www.geeksforgeeks.org/introduction-to-algorithms/)__ 

Following the asymptotic notations to calculate the running time complexity of an algorithm, we have 3 types of running time for algorithms:
<br>
<br>
* Big-Ο Notation or the worst case which is the upper bound of an algorithm's running time and it's mathematical expression can be described as: 
<br>
<br>
O(g(n)) = { f(n): there exist positive constants c and n0
            such that 0 ≤ f(n) ≤ cg(n) for all n ≥ n0 }
<br>
<br>
In this case, the running time of the algorithm does not cross the the time provided by 0(g(n)), for any value of n.
<br>
<br>
* Omega notation which is the lower bound of an algorithm's running time and it provides the best case complexity of an algorithm. The mathematical expression of the Ω-notation can be expressed as below:
<br>
<br>
Ω(g(n)) = { f(n): there exist positive constants c and n0 
            such that 0 ≤ cg(n) ≤ f(n) for all n ≥ n0 }
<br>
<br>
Here, the Ω(g(n)) gives the minimum time required by the algorithm for any value of n.
<br>
* Theta notation which encloses both the lower and upper bound of an algorithm's running time is used for analysing the average case complexity of an algorithm. From the mathematical point of view, if the function f(n) lies anywhere between c1g(n) and c2g(n) for all n ≥ n0, then f(n) is said to be asymptotically tight bound.
<br>
<br>
Θ(g(n)) = { f(n): there exist positive constants c1, c2 and n0
            such that 0 ≤ c1g(n) ≤ f(n) ≤ c2g(n) for all n ≥ n0 }
<br>

## 4.2 Binary Search __[medium](https://medium.com/techie-delight/binary-search-practice-problems-4c856cd9f26c)__

Binary search is a Divide & Conquer algorithm which means to find an element in previously sorted array, we need to divide the array in two subarrays and iteratively operating on subarrays. The binary search  will  define the search space by using the property of the array that it is sorted. 
<br>
<br>
This can be done by comparing the mid-value in the search space to the target value. If the target value matches the middle element,its position in array is returned. Otherwise, half of the search space will be discarded. The example below illustrates the implementation of binary search: 

```python
# Function to determine if a `target` exists in the sorted list `nums`
# or not using a binary search algorithm
def binarySearch(nums, target):
 
    # search space is nums[left…right]
    (left, right) = (0, len(nums) - 1)
 
    # loop till the search space is exhausted
    while left <= right:
 
        # find the mid-value in the search space and
        # compares it with the target
 
        mid = (left + right) // 2
 
        # overflow can happen. Use:
        # mid = left + (right - left) / 2
        # mid = right - (right - left) // 2
 
        # target is found
        if target == nums[mid]:
            return mid
 
        # discard all elements in the right search space,
        # including the middle element
        elif target < nums[mid]:
            right = mid - 1
 
        # discard all elements in the left search space,
        # including the middle element
        else:
            left = mid + 1
 
    # `target` doesn't exist in the list
    return -1
 
 
if __name__ == '__main__':
 
    nums = [2, 5, 6, 8, 9, 10]
    target = 5
 
    index = binarySearch(nums, target)
 
    if index != -1:
        print('Element found at index', index)
    else:
        print('Element found not in the list')
```

The time complexity of the binary search is O(logn) which is very efficient as the search space is reduced to half after each iteration. This can be figured out by this simple mathematic operation:

##### Example: 
How to implement a function recursively to get the desired Fibonacci sequence value?

```python
def get_fib(position):
    if position == 0 or position == 1:
        return position
    return get_fib(position - 1) + get_fib(position - 2)
# Test cases
print (get_fib(9))
print (get_fib(11))
print (get_fib(0))
```

## 5.1 Bubble Sort

In bubble sort, we have to compare the elements of the array side by side and switch them two by two till we sort them in a desired(ascending or descending)way.
<br>
For instance, we can implement the naive approach of buble sort and figure out the number of comparisons in the array below like:

```python
arr = [7,3,2]
```

Start comparing 7 > 3 (1st comparison & switch)--> 7 > 2 (2nd comparison & switch) --> number of comparison for the first round is 2 which is one less than the size of the array.The result arrays will look like :

```python
arr = [3,2,7]
```

Start comparing 3 > 2 (1st comparison & switch) --> 3 < 7 (2nd comparison & no switch)
<br>
As can be seen it always takes us one less than the number of elements in the array for the comparison. So the runtime is n-1 which will be 2 here.

## 5.1.1 Efficiency of bubble sort

As we saw in the example above for n-1 iterations, we had n-1 comparisons which leads us to the time complexity of O(n^2).
<br>
In the avarage case,in each iteration when lower numbers are bubbled, the last two numbers were unchanged and comparison was not meaningful. Therefore, the average time complexity of the bubble sort is O(n^2). 
<br>
In the best case, if there is only one number that needs to be sorted towards the upper bound or if the array is already sorted and we need only one iteration to go through it, the time complexity will be O(n).

The below example illustrates the implementation of bubble sort in python: __[Geeks for geeks](https://www.geeksforgeeks.org/bubble-sort/)__

```python
def bubbleSort(arr):
    n = len(arr)
 
    # Traverse through all array elements
    for i in range(n):
 
        # Last i elements are already in place
        for j in range(0, n-i-1):
 
            # traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]
 
# Driver code to test above
arr = [64, 34, 25, 12, 22, 11, 90]
 
bubbleSort(arr)
 
print ("Sorted array is:")
for i in range(len(arr)):
    print ("%d" %arr[i]),
```

## 5.1.2 Space complexity of Bubble sort

Since we don't need any added data structures or array or extra storage, the space complexity is constant O(1) and bubble sort is considered as one of the in-place sorting algorithms.

### Example: __[Geeksfor geeks](https://practice.geeksforgeeks.org/problems/merge-sort/1)__
<br>
Given an array arr[], its starting position l and its ending position r. Sort the array using merge sort algorithm.

```python
class Solution:    
    def merge(self,arr, l, m, r): 
        #finding the mid of the array
        n1 = m - l + 1
        n2 = r - m 
        
        #dividing the array elements into 2 halves
        L = [0] * (n1) 
        R = [0] * (n2) 
        
        #sorting the first half
        for i in range(0 , n1): 
            L[i] = arr[l + i] 
      
        #sorting the second half
        for j in range(0 , n2): 
            R[j] = arr[m + 1 + j] 
      
        i = 0     # Initial index of first subarray 
        j = 0     # Initial index of second subarray 
        k = l     # Initial index of merged subarray 
      
        #copying data to temporary arrays L[] & R[] and checking 
        #if any element was left
        while i < n1 and j < n2 : 
            if L[i] <= R[j]: 
                arr[k] = L[i] 
                i += 1
            else: 
                arr[k] = R[j] 
                j += 1
            k += 1
      
        while i < n1: 
            arr[k] = L[i] 
            i += 1
            k += 1
      
        while j < n2: 
            arr[k] = R[j] 
            j += 1
            k += 1
            
    def mergeSort (self, arr, l, r):
        if l < r:
            m = l + (r - l)//2
            
            self.mergeSort (arr, l, m)
            self.mergeSort (arr, m+1, r)
            self.merge (arr, l, m, r)
```

## 5.3 Quick Sort __[Khanacademy](https://www.khanacademy.org/computing/computer-science/algorithms/quick-sort/a/overview-of-quicksort#:~:text=Like%20merge%20sort%2C%20quicksort%20uses,from%20how%20merge%20sort%20does.)__

Quick sort can be considered as one of the most efficient sorting algorithms which uses divide and conquer and is a recursive algorithm. To do the quick sort, we randomly choose a value from the array(which is called pivot)and the elements which are less or equal to the pivot are arranged to the left side of the array and the elemnts which are greater than the pivot are positioned to its right.
<br>
This process is called partitioning and we will proceed till all the elemnts are sorted in the array.

### 5.3.1 Quick Sort implementation in Python __[Geeks](https://www.geeksforgeeks.org/quick-sort/)__

```python
def partition(start, end, array):
      
    # Initializing pivot's index to start
    pivot_index = start 
    pivot = array[pivot_index]

    while start < end:
          
        while start < len(array) and array[start] <= pivot:
            start += 1
    
        while array[end] > pivot:
            end -= 1

        if(start < end):
            array[start], array[end] = array[end], array[start]
      
    array[end], array[pivot_index] = array[pivot_index], array[end]
     
    return end
      
# The main function that implements QuickSort 
def quick_sort(start, end, array):
      
    if (start < end):

        p = partition(start, end, array)
          
        quick_sort(start, p - 1, array)
        quick_sort(p + 1, end, array)
          
# Driver code
array = [ 10, 7, 8, 9, 1, 5 ]
quick_sort(0, len(array) - 1, array)
  
print(f'Sorted array: {array}')
```

### 5.3.2 Efficiency of Quick Sort

Worst case: Every time the array is divided into half the amount of comparisons are cut. In case, we wouldn't be able to cut the array into halves, we have to do all the comparisons every time. There are some scenarios when the pivots are already in their correct places and we have to do all the comparisons each time which resembles the bubble sort efficiency O(n^2).
<br>
Best case: When the pivot is moved to the middle and we split the array into halves everytime, the other arrays' pivot can be moved to the middle and the sorting can be done very quickly and the efficiency is simillar to merge sort.

### 5.3.3 Optimization on Quick Sort

1. Configuring the program in a way that consumes the same computing powewr with less amount of time that runs the halves parallelly.
<br>
2. Instead of selecting the last element as pivot, selecting the median of them as a pivot will give a better idea of what elemnts are in the array and moving the element to the middle is more probable and have a best case scenario.

### 5.3.4 Space complexity of Quick Sort

Since no extra space is being used, the space complexity is constant O(1).

Add the three functions of "get_position" , "insert" & "delete" to the LinkedList class.

```python
class Element(object):
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList(object):
    def __init__(self, head=None):
        self.head = head

    def append(self, new_element):
        current = self.head
        if self.head:
            while current.next:
                current = current.next
            current.next = new_element
        else:
            self.head = new_element

    def get_position(self, position):
        current = self.head
        position_counter = 1
        if position < 1:
            return None
        while True:
            if position_counter == position:
                return current
            else:
                position_counter += 1
            if current.next == None:
                break
            current = current.next    
        return None

    def insert(self, new_element, position):
        current = self.head
        position_counter = 1
        while True:
            if position_counter == position-1:
                previous = current
                current = previous.next
                break
            current = current.next
            position_counter += 1
        previous.next = new_element
        new_element.next = current  
        return None
    
    def delete(self, value):
        current = self.head
        previous = None
        while current.value != value and current.next:
            previous = current
            current = current.next
        if current.value == value:
            if previous:
                previous.next = current.next
            else:
                self.head = current.next



# Test cases
# Set up some Elements
e1 = Element(1)
e2 = Element(2)
e3 = Element(3)
e4 = Element(4)

# Start setting up a LinkedList
ll = LinkedList(e1)
ll.append(e2)
ll.append(e3)

# Test get_position
# Should print 3
print (ll.head.next.next.value)
# Should also print 3
print (ll.get_position(3).value)

print("")

# Test insert
ll.insert(e4,3)
# Should print 4 now
print (ll.get_position(3).value)

print("")

# Test delete
ll.delete(1)
# Should print 2 now
print (ll.get_position(1).value)
# Should print 4 now
print (ll.get_position(2).value)
# Should print 3 now
print (ll.get_position(3).value)
```

```python
class Element(object):
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList(object):
    def __init__(self, head=None):
        self.head = head

    def append(self, new_element):
        current = self.head
        if self.head:
            while current.next:
                current = current.next
            current.next = new_element
        else:
            self.head = new_element

    def insert_first(self, new_element):
        new_element.next = self.head
        self.head = new_element

    def delete_first(self):
        ref = self.head
        if self.head:
            self.head = self.head.next
        return ref

class Stack(object):
    def __init__(self,top=None):
        self.ll = LinkedList(top)

    def push(self, new_element):
        self.ll.insert_first(new_element)

    def pop(self):
        return self.ll.delete_first()

# Test cases
# Set up some Elements
e1 = Element(1)
e2 = Element(2)
e3 = Element(3)
e4 = Element(4)

# Start setting up a Stack
stack = Stack(e1)

# Test stack functionality
stack.push(e2)
stack.push(e3)
print(stack.pop().value)
print(stack.pop().value)
print(stack.pop().value)
print(stack.pop())
stack.push(e4)
print(stack.pop().value)
```

```python
class Queue(object):
    def __init__(self, head=None):
        self.storage = [head]

    def enqueue(self, new_element):
        self.storage.append(new_element)

    def peek(self):
        return self.storage[0]

    def dequeue(self):
        return self.storage.pop(0)

# Setup
q = Queue(1)
q.enqueue(2)
q.enqueue(3)

# Test peek
# Should be 1
print (q.peek())

# Test dequeue
# Should be 1
print(q.dequeue())

# Test enqueue
q.enqueue(4)
# Should be 2
print(q.dequeue())
# Should be 3
print(q.dequeue())
# Should be 4
print(q.dequeue())
q.enqueue(5)
# Should be 5
print(q.peek())
```

## 6.10 Hashing

Hash functions let us implement look ups in constant time while a list or set does it in linear time. This ability of hash functions make their runtime super fast. What these functions can do behind the scene is that they convert some values to one that can be redeemed more easily and the outcome of the function is the index in an array.
<br>
A popular strategy in hash functions is to divide the last few digits(as they are the most random ones) by a specific number and using the remainder to be as an index to store that value in the array.
<br>
This strategy will enable us to get the value we want in constant time.

## 6.11 Collisions

However, hash functions do not work perfectly all the time and there are possibilities that they produce the same remainder for different values. This condition is known as Collision.

### 6.11.1 How to prevent Collisions?

There are 2 major approaches to prevent collisions:
<br>
1. Changing the value in hash function or the function itself: give us more place to keep our values
<br>
2. Changing the structure of the array: Using kind of lists called buckets which contain more than one hashed value in one place

### 6.11.2 The drawbacks of the approaches above

1. In the first approach we can have a constant time look up O(1), but bigger number as a hash function needs more memory to store the values. Also, continous change in the hash function value and moving array values will add on time & size complexity.
<br>
2. On average, hash functions have O(1) time complexity but with bucket system we might end up iterating over a list and end up having O(m) runtime(m= number of elements in a list).
<br>
In the end, there is always a trade-off between runtime & space complexity in every approach we try to implement.

## 6.12 Load Factor  __[Udacity](https://classroom.udacity.com/courses/ud513/lessons/7118294395/concepts/79336691870923)__

One good indicator which help us to figure out whether our hash table is optimal or not is to calculate the "Load Factor" which can be defined as below: 
<br>
Load Factor = Number of Entries / Number of Buckets
<br>
A load factor close to 1 represents that we have a full hash table and the number of values are equal to the number of buckets and the chances of having collisions are high. On the other hand, a load factor close to 0 means that most of our buckets are empty and we need to rehash or use less buckets to not waste our memory storage.

## 6.15 Trees

2. Every element in a tree which contain data is called node and the nodes must be connected.
<br>
3. We shouldn't enter a node twice (cycles are not allowed)
<br>
4. The root has level one and it is a parent for the following nodes which are children with descending levels. However, based on how many levels we have some of the nodes can be both parent and child.

5. Children with the same parent are siblings to each other and they are the descendant of the root node which is their ancestor.
<br>
6. Nodes with no children are called leaves and external while the parent is intertnal.
<br>
7. What connects two nodes is called an edge and the collection of the edges forms a path.
<br>
8. Hight of a node can be described as the number of edges between itself and furthest leaf.
<br>
9. Height of a tree is height of the root node.
<br>
10. The number of edges to the root is the depth of the node and the height & depth of the node proceed oppositely.


### 6.15.1 Tree Traversal __[Udacity](https://classroom.udacity.com/courses/ud513/lessons/7122604912/concepts/71162844790923)__

Since trees are not linear, it is not very easy to walk through them as easy as lists and it is important to traverse all the nodes when searching.

### 6.15.2 Main approaches in Tree Traversal

1. Depth-first search(DFS): The priority is to visit the children nodes first. 
<br>
2. Breadth-first search(BFS): Exploring same level nodes first before going to child nodes.

### 6.15.3 Different Depth-First Traversal

There are various ways to do DFS as following:
<br>
1. Pre-order: Starting from the route and choosing one of the children (left by convention) and crossing down the nodes till we reach to a leaf. After checking the leaf , we go to its parent to check and then we go to its child. When one sub tree is completed we go back to the root and check the right child and so on.
<br>
2. In-order: Going through the nodes in order is the key here. Traversing to the left child till reaching to a leaf and then going back to its parent and then its child and we need to continue this in order till we finish all nodes. In this case we walk through the nodes pretty much from left to right.
<br>
3. Post-order: The intention is to check all of a node decendants before moving to the next node.
